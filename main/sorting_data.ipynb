```python
import glob
import pandas as pd
import os
import re
```

```python
files = glob.glob("/Users/benediktas/dev/University/data-science/tennis/tennis_atp/atp_matches_*.csv")

selected_files = []

for name in files:
    match = re.search(r'atp_matches_(\d{4})\.csv', name)

    if match:
        year = int(match.group(1))
        if 1985 < year:
            selected_files.append(name)
```

```python
dfs = []
for file in selected_files:
    dfs.append(pd.read_csv(file))

df = pd.concat(dfs)

df['tourney_date'] = pd.to_datetime(df['tourney_date'], format='%Y%m%d', errors='coerce')
df = df.sort_values("tourney_date")
```
